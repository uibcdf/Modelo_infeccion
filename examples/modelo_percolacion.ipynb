------------------------------------

```python
n_filas = 100
n_columnas = 200
ocupaciones = np.linspace(1.0, 0.0, num=30)
n_realizaciones = 100
lista_porcentaje_infectados_promedio = []

for ocupacion in ocupaciones:
    print(ocupacion)
    lista_porcentaje_infectados = []

    for _ in range(n_realizaciones):
    
        sistema = mi.creacion_sistema(n_filas,n_columnas,ocupacion)
        mi.infeccion_inicial(sistema)
        mi.propagacion(sistema)
        p = mi.porcentaje_infectado(sistema)
        lista_porcentaje_infectados.append(p)
    
    m = np.mean(lista_porcentaje_infectados)
    lista_porcentaje_infectados_promedio.append(m)
```

```python
plt.plot((1-np.array(ocupaciones))*100,np.array(lista_porcentaje_infectados_promedio)*100)
plt.axvline(x=59, ymin=0, ymax=1, color="red")
plt.ylabel("% Infectados")
plt.xlabel("% en confinamiento")
plt.show()
```
